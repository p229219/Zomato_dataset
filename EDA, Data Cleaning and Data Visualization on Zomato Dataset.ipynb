#            Data Pre-processing

```python
#deleting the not required columns
df.drop(['url','dish_liked','phone',],axis=1,inplace=True)
```

```python
df.head()
```

```python
#checking for duplicate entries
df.duplicated().sum()
```

```python
#removing the duplicate entries
df.drop_duplicates(inplace=True)
```

```python
#cross checking for duplicate entries
df.duplicated().sum()
```

```python
#after removing duplicates checking for no.rows and columns
df.shape
```

```python
#missing values
[features for features in df.columns if df[features].isnull().sum()>1]
```

```python
#checking for nan value
df.isnull().sum()
```

```python
#removing the nan value
df.dropna(how="any",inplace=True)
```

```python
#rechecking the nan value present or not
df.isnull().sum()
```

```python
df.info()
```

```python
df.columns
```

```python
#renaming the columns name
df=df.rename(columns={'approx_cost(for two people)':'cost','listed_in(type)':'type',
                                  'listed_in(city)':'city'})
df.columns
```

```python
df["cost"].dtype
```

```python
df.head()
```

```python
#converting cost into string
df["cost"]=df["cost"].astype(str)
#Using lambda function to replace ',' from cost
df['cost']=df['cost'].apply(lambda x: x.replace(',','.'))
# Changing the cost to Float
df["cost"]=df["cost"].astype(float)
df.info()
```

```python
#reading unique values from rate columns
df["rate"].unique()
```

```python
#Removing '/5' from Rates
df=df.loc[df.rate !='NEW']
df=df.loc[df.rate !='-'].reset_index(drop=True)
df_slash=lambda x: x.replace('/5', '') if type(x) == np.str else x
df.rate=df.rate.apply(df_slash).str.strip().astype('float')
#checking the dataset after transformation
df['rate'].head()
```

```python
# Adjust the column nam
df.name=df.name.apply(lambda x:x.title())
df.online_order.replace(('Yes','No'),(True, False),inplace=True)
df.book_table.replace(('Yes','No'),(True, False),inplace=True)
df.head() # looking at the dataset after transformation
```

```python
#checking for unique cost
df["cost"].unique
```

```python
#Encoding the input Variables
def Encode(df):
    for column in df.columns[~df.columns.isin(['rate', 'cost', 'votes'])]:
        df[column] = df[column].factorize()[0]
    return df

df_en = Encode(df.copy())
df_en.head()
```

```python
#Correlation between different variables
corr = df_en.corr(method='spearman')
plt.figure(figsize=(15,8))
sns.heatmap(corr, annot=True)
df_en.columns
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
from sklearn.metrics import r2_score
```

```python
#the independent variables and dependent variables
x = df_en.iloc[:,[2,3,5,6,7,8,9,11]]
y = df_en['rate']
#Test and Training Set
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.1,random_state=353)
x_train.head()
```

```python
#Decision Tree Regression
from sklearn.tree import DecisionTreeRegressor
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.1,random_state=105)
DTree=DecisionTreeRegressor(min_samples_leaf=.0001)
DTree.fit(x_train,y_train)
y_predict=DTree.predict(x_test)
from sklearn.metrics import r2_score
r2_score(y_test,y_predict)
```

```python
#Random Forest REgression
from sklearn.ensemble import RandomForestRegressor
RForest=RandomForestRegressor(n_estimators=500,random_state=329,min_samples_leaf=.0001)
RForest.fit(x_train,y_train)
y_predict=RForest.predict(x_test)
from sklearn.metrics import r2_score
r2_score(y_test,y_predict)
```
